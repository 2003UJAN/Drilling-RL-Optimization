**Install Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Random seed for reproducibility**

```python
np.random.seed(42)
```

**Number of samples**

```python
num_samples = 10000
```

**Generate synthetic drilling parameters**

```python
torque = np.random.uniform(30, 70, num_samples)  # Nm
pressure = np.random.uniform(1000, 5000, num_samples)  # PSI
speed = np.random.uniform(50, 200, num_samples)  # RPM
```

**Efficiency function**

```python
efficiency = np.clip(1 - (np.abs(torque - 50) * 0.01 + (pressure - 3000) * 0.0002), 0, 1)
```

**Wear factor**

```python
wear = np.clip((np.abs(torque - 50) * 0.02 + (pressure - 3000) * 0.0005), 0, 1)
```

**Create DataFrame**

```python
df = pd.DataFrame({
    "Torque (Nm)": torque,
    "Pressure (PSI)": pressure,
    "Speed (RPM)": speed,
    "Efficiency": efficiency,
    "Wear Factor": wear
})
```

**Save dataset**

```python
df.to_csv("synthetic_drilling_data.csv", index=False)
print("Synthetic dataset saved as 'synthetic_drilling_data.csv'!")
```

**Plot data distribution**

```python
sns.pairplot(df, diag_kind="kde")
plt.show()
```
